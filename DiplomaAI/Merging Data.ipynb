```python
import pandas as pd
from datetime import datetime
```

```python
def Merge_Commodtites(name):
    
    Trade_df = pd.read_excel(f'../Data/New_Work(Revisit_Model)/trade_data/{name}_TradeData_21-22.xlsx')
    # Merge 'year' and 'month' columns and create a new 'date' column
    Trade_df['Date'] = pd.to_datetime(Trade_df[['Year', 'Month']].assign(day=1))
    Trade_df.drop(columns=['Year', 'Month'], inplace=True)
    Trade_df = Trade_df.iloc[:,1:]
#     print(Trade_df.shape)
#     FAO_df = pd.read_excel(f'../Data/New_Work(Revisit_Model)/FAO-Production-Yield-Area harvested.xls',sheet_name=f'{name}')
#     FAO_df.rename(columns={'Year': 'Date'}, inplace=True)
#     print(FAO_df.shape)
    Price_df = pd.read_excel(f'../Data/New_Work(Revisit_Model)/prices2020-2023.xlsx',sheet_name=f'{name}')
#     print(Price_df.tail())
#     print(Price_df.shape)
    Prod_df = pd.read_csv(f'../Data/New_Work(Revisit_Model)/Prod_data/Prod_{name}.csv')
    Prod_df.rename(columns={'Year': 'Date'}, inplace=True)
    Prod_df.drop(columns=['Imports (t)','Exports (t)','Exports per capita (kg)','exports__tonnes__per_capita',
                         'Imports per capita (kg)','imports__tonnes__per_capita','Date'],inplace=True, errors='ignore')
    Prod_df = Prod_df.add_prefix('Prod ')
#     Prod_df.columns = [col.replace('(t)','Kg') for col in Prod_df.columns]
    Prod_df['Date'] = pd.date_range(start=datetime(2010, 1, 1), periods=Prod_df.shape[0], freq=pd.DateOffset(years=1))

    Data_m = pd.DataFrame()
    Data_m['Date'] = pd.date_range(start='01/01/2010', end='01/07/2023', freq='MS')
    
    #Political and Social
    Political_df = pd.read_excel(f'../Data/New_Work(Revisit_Model)/Political/Political_and_social_volatilities-EU_2013_2022.xlsx',sheet_name='Indicators')
    
    #Read Financial Data
    Finicial = pd.read_excel(f'../Data/New_Work(Revisit_Model)/Financial_Data/CPI-GDP-2013-2023.xlsx')
    
    # Merge all DataFrames on the 'Date' column
    merged_df = Prod_df.merge(Data_m, on='Date', how='outer')\
                .merge(Political_df , on='Date', how='outer')\
                .merge(Trade_df, on='Date', how='outer')\
                .merge(Price_df , on='Date', how='outer') #                 .merge(FAO_df, on='Date', how='outer')\
    merged_df.sort_values(by='Date',inplace=True)
#     merged_df.ffill(axis = 0,inplace=True)
    merged_df['Year'] = merged_df['Date'].dt.year
    merged_df['Month'] = merged_df['Date'].dt.month
    merged_df = merged_df.merge(Finicial , on=['Year','Month'], how='outer')
    merged_df.to_excel(f'../Data/New_Work(Revisit_Model)/Final_Data/{name}_Data_10_23.xlsx')
    
    return  merged_df
```
